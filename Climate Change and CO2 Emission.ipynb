### Data Polishing & Cleaning 

```python
# Exclude useless columns 
co2_to_use = co2.drop(columns=['Series Name', 'Series Code', 'Country Code'])

# Check new df structure
print(co2_to_use.columns)
```

```python
# Change column names (years)
co2_to_use.columns = co2_to_use.columns[:1].tolist() + co2_to_use.columns[1:].str.extract(r'(\d{4})')[0].tolist()

# Check columns names
print(co2_to_use.columns)
```

```python
# Replace '..' with nan
co2_dot_to_nan = co2_to_use.replace('..', np.nan)
print(co2_dot_to_nan.columns)
```

```python
# Check NaNs by columns 
print(co2_dot_to_nan.isnull().sum(axis=0))
```

```python
# Exclude columns 2020 and 2021 as it has only NaNs values 
co2_columns = co2_dot_to_nan.drop(columns = ['2020','2021'])
print(co2_columns.columns)
```

```python
# Drop NaNs
co2_clean = co2_columns.dropna()

# Check nans
print(co2_clean.isnull().sum())
```

```python
# Check duplicated 
print(co2_clean.duplicated().sum())
```

```python
# Check data shape 
print(f'Old df shape: {co2.shape}')
print(f'New df shape: {co2_clean.shape}')
```

```python
# Create a new df 
co2_final = co2_clean.copy()

# Select all columns to be changed 
cols = co2_final.columns[1:]

# Change data types
co2_final[cols] = co2_final[cols].astype(float)

# Check the result 
print(co2_final.dtypes)
```

```python
# Round values 
co2_rounded = co2_final.round(0)

print(co2_rounded)
```

### Data Vizualization

```python
# Drop 'World' row form the df 

# Set Country Name as index 
co2_rounded.set_index('Country Name', inplace = True)

# Drow 'World' column
co2_rounded.drop(index = 'World', inplace = True)

# Reset the index 
co2_rounded.reset_index(inplace= True)
```

```python
# Total CO2 emission by country over all years 

# New df 
co2_sum = co2_rounded.copy()

# Set Country Name as index 
co2_sum.set_index('Country Name', inplace = True)

# Sum co2 emission for each country 
co2_sum['Total CO2 emission'] = co2_sum.sum(axis =1)

# Sum all emissions 
total_emissions = co2_sum['Total CO2 emission'].sum()

# Total emission percentage by country 
co2_sum['% Emission from Total'] = round((co2_sum['Total CO2 emission']/ total_emissions)*100,2)

# Sort the countries by CO2 emission
co2_sum = co2_sum.sort_values('% Emission from Total', ascending = False)

# Select the top 5 (sum)
head_5_percentage = co2_sum.head(5)

# Select the country with less CO2 emission
tail_5_percentage = co2_sum.tail(5)

# Select % Emission from Total for the head 10 
head_result = head_5_percentage.loc[:,['% Emission from Total']]

# Select % Emission from Total for the tail 10
tail_result =  tail_5_percentage.loc[:,['% Emission from Total']]

# Reset index
head_result.reset_index(inplace=True)

# Plot a bar chart 
head_result.plot(kind='bar', x='Country Name', y='% Emission from Total', figsize=(10,8))
plt.title('Top 5 CO2 Emitters')
plt.xlabel('Country')
plt.ylabel('% Emission from Total')
plt.show()
```

```python
# Line chart for each country carbon production

# New df
co2_line = co2_sum.copy()

# Reset index
co2_line.reset_index(inplace = True)

# Figure size 
fig, ax = plt.subplots(figsize=(25, 16))

# Plot line chart
for country in co2_line['Country Name']:
    plt.plot(co2_line.columns[1:31], co2_line.loc[co2_line['Country Name']==country].values[0][1:31], label=country)

plt.legend()
plt.title('CO2 Emission by Country Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emission')
plt.show()
```

```python
# Creating a new csv from co2_sum df 
co2_emission = co2_sum.copy()

co2_emission.reset_index(inplace= True)

co2_emission.to_csv('co2_emission.csv', index = False)
```

### Data Polishing & Cleaning 

```python
# Exclude useless columns 
climate_to_use = climate.drop(columns=['ObjectId', 'Unit', 'Change '])

# Check new df structure
print(climate_to_use.columns)
```

```python
# Set the year range the equal to co2 dataset year range 

# Set start year
start_year = 1990

# Set end year 
end_year = 2019

# Slice the df to match the CO2 years range
climate_year_range = climate_to_use.loc[:, ['Country Name'] + [str(year) for year in range(start_year, end_year+1)]]

print(climate_year_range)
```

```python
# Exclude countries that has 60% or more missing data (18 missing values)

# New df 
climate_no_nan = climate_year_range.copy()

# Count the number of missing values in each row
missing_values = climate_no_nan.isnull().sum(axis=1)

# Filter out the rows with more than 60% missing values
climate_clean = climate_no_nan[missing_values <= 18]

# Data shape
print(climate_clean.shape)

# Check remaining NaNs
print(climate_clean.isnull().sum().sum())
```

```python
# Fill the missing data with the year mean 

# Copy the original data to a new one
climate_mean = climate_clean.copy()

# Convert columns to numeric
climate_mean.iloc[:, 1:] = climate_mean.iloc[:, 1:].apply(pd.to_numeric)

# Calculate column means
col_means = climate_mean.iloc[:, 1:].mean(axis=0)

# Iterate over columns and fill missing values with column means
for column in climate_mean.columns[1:]:
    # Get column index
    col_idx = climate_mean.columns.get_loc(column)
    # Calculate column mean
    col_mean = col_means[col_idx-1]
    # Fill missing values with column mean
    climate_mean[column] = climate_mean[column].fillna(col_mean)

# check for NaN values
print(climate_mean.isnull().sum().sum())
```

```python
# Group the three China's region into one region names China 

# New df 
china_as_one = climate_mean.copy()

# Extract the country name from the "Country Name" column
china_as_one['Country First Name'] = china_as_one['Country Name'].apply(lambda x: x.split(':')[0])

# Group by the country name and calculate the mean values
china_as_one = china_as_one.groupby('Country First Name').apply(lambda x: x.mean())

# Reset the index to convert the "Country" index to a column
china_as_one = china_as_one.reset_index()

# Rename the Country First Name to Country Name again. 
china_as_one = china_as_one.rename(columns={'Country First Name': 'Country Name'})

# Rename China, P.R. to China 
china_as_one['Country Name'] = china_as_one['Country Name'].apply(lambda x: 'China' if x == 'China, P.R.' else x)
```

### Data Visualization

```python
# Line chart for each country climate change over time

# New df
climate_line = china_as_one.copy()

# Figure size 
fig, ax = plt.subplots(figsize=(25, 16))

for country in climate_line['Country Name']:
    plt.plot(climate_line.columns[1:], climate_line.loc[climate_line['Country Name']==country].values[0][1:], label=country)

ax.axhline(y=0, color='black', linestyle='--', label='Grade line')
plt.legend()
plt.title('Temperature changes in all countries over time')
plt.xlabel('Year')
plt.ylabel('Temperature change')
plt.show()
```

```python
# Most and less affected countries by the climate change and the world comparetion

# New df 
climate_affected = climate_line.copy() 

# Calculate average values for each country
averages = climate_affected.mean(axis=1)
averages.index = climate_affected['Country Name']

# Sort values 
averages = averages.sort_values()

# Sort by most and less affected 
most_aff = averages.tail(3)
least_aff = averages.head(3)

# Show results 
print(' ')
print('Most affected Countries')
print(most_aff)
print(' ')
print('Least affected Countries')
print(least_aff)

# Exclude 'Pitcairn Islands' as it will not be considered a Country 
# Select rows where the 'Country Name' column equals 'Pitcairn Islands'
index_to_drop = climate_affected[climate_affected['Country Name'] == 'Pitcairn Islands'].index

# Drop the selected rows
climate_countries = climate_affected.drop(index_to_drop)

# Calculate average values for each country
averages_climate_countries = climate_countries.mean(axis=1)
averages_climate_countries.index = climate_countries['Country Name']

# Sort values 
averages_climate_countries = averages_climate_countries.sort_values()

# Sort by most and less affected 
most_aff_climate_countries = averages_climate_countries.tail(5)
least_aff_climate_countries = averages_climate_countries.head(5)

# Show Results
print(' ')
print('Most and less afected countries filtered')
print(' ')
print('Most affected Countries')
print(most_aff_climate_countries)
print(' ')
print('Least affected Countries')
print(least_aff_climate_countries)
```

```python
# Set figure size 
fig, ax = plt.subplots(figsize=(25, 16))

# Plot the data for the most and least affected countries
for country in most_aff_climate_countries.index:
    plt.plot(climate_affected.loc[climate_affected['Country Name']==country].iloc[:,1:].values[0], label=country, linestyle='-', linewidth=1)

for country in least_aff_climate_countries.index:
    country_data = climate_affected.loc[climate_affected['Country Name']==country].iloc[:,1:].values
    if len(country_data) > 0:
        plt.plot(country_data[0], label=country, linestyle='--', linewidth=2)

# Plot the line change for the world data 
plt.plot(climate_countries.loc[climate_countries['Country Name']=='World'].iloc[:,1:].values[0], label='World', linestyle='-', linewidth=10, alpha=0.5, color = 'blue')

# Set the x-axis ticks and labels to be the years in the dataset
years = climate_countries.columns[1:]
plt.xticks(range(len(years)), years, rotation=45)

# Set labels and legends
ax.axhline(y=0, color='black', linestyle='--', label='Zero mark')
plt.legend()
plt.title("Comparison of changes over time among the most and least affected countries, as well as the world")
plt.xlabel('Year')
plt.ylabel('Change')
plt.show()
```
